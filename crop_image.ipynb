```python
import cv2
from matplotlib import pyplot as plt

path_image ='original/images/'
path_mask = 'original/masks/'

file_name = open('image_name.txt', 'r')
lines = file_name.readlines()
file = lines[68].strip()
print('File name :',file)

img = cv2.cvtColor(cv2.imread(path_image + file), cv2.COLOR_BGR2RGB)
mask = cv2.cvtColor(cv2.imread(path_mask + file), cv2.COLOR_BGR2RGB)

print('image shape :',img.shape) # Print image shape
print('mask shape :',mask.shape) # Print image shape

crop_size_height,crop_size_width = 1088,1920

cropped_image = []
cropped_mask = []
for i in range(2):
    x_start, y_start= 100 + (i*1088), 1300
    x_end = x_start + crop_size_height
    y_end = y_start + crop_size_width
    cropped_image.append(img[x_start:x_end, y_start:y_end])
    cropped_mask.append(mask[x_start:x_end, y_start:y_end])

# print(len(cropped_image))
plt.figure(figsize=(15,20))
plt.subplot(5,2,1) 
plt.title('original image')
plt.imshow(img)
plt.subplot(5,2,2)
plt.title('original mask')
plt.imshow(mask)
ax = 3
for i in range(len(cropped_image)):
    print(cropped_image[i].shape)
    print(cropped_mask[i].shape)
    plt.subplot(5,2,ax)
    plt.title('cropped img '+str(i))
    plt.imshow(cropped_image[i])
    plt.subplot(5,2,ax+1)
    plt.title('cropped mask '+str(i))
    plt.imshow(cropped_mask[i])
    ax += 2

plt.show()
```

```python
# Save the cropped image
for i in range(len(cropped_image)):
    name = file[:-4] + '_' + str(i) + '.png'
    cv2.imwrite(f"cropped/images/{name}", cropped_image[i])
    cv2.imwrite(f"cropped/masks/{name}" , cropped_mask[i])
```
